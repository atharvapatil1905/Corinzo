### **4. SELECT MIN(year), AVG(year), MAX(year), engine, manufacturer FROM planes GROUP BY engine, manufacturer**

```python
planes_df.head(5)
```

```python
# Pandas solution
task5_my = planes_df.groupby(by=['engine', 'manufacturer'])['year'].agg(['min', 'mean', 'max']).reset_index()
task5_my.head()
task5_my.rename(columns={'min': 'MIN(year)', 'mean': 'AVG(year)',
                         'max': 'MAX(year)'}, inplace=True)
task5_my
```

### **6. SELECT tailnum FROM planes WHERE seats BETWEEN 150 AND 190 AND year >= 2012**

```python
planes_df.head(5)
```

```python
# Pandas solution
task7_my = planes_df[(planes_df['year']>=2012) & (planes_df['seats']>150) & (planes_df['seats']<190)]['tailnum']
task7_my = task7_my.reset_index(drop=True)
# task7_my = pd.DataFrame(task7_my)
task7_my
```

### **7. SELECT tailnum, manufacturer, seats FROM planes WHERE manufacturer IN ("BOEING", "AIRBUS", "EMBRAER") AND seats>390**

```python
planes_df.head()
```

```python
# Pandas solution
list_manufacturer = ["BOEING", "AIRBUS", "EMBRAER"]
task8_my = planes_df[(planes_df['manufacturer'].isin(list_manufacturer)) & (planes_df['seats']>390)]
task8_my = task8_my[['tailnum','manufacturer','seats']]
task8_my.reset_index(drop=True, inplace=True)
task8_my
```

### **9. Give me top 5 manufacturers**

```python
planes_df.head()
```

```python
# Pandas solution
task9_my = planes_df.groupby('manufacturer')['tailnum'].count()
task9_my = pd.DataFrame(task9_my)
# task9_my
task9_my.reset_index(inplace=True)
task9_my.rename(columns={'tailnum':'count'}, inplace=True)
task9_my = task9_my.sort_values(by='count', ascending=False).head(5)
task9_my
```
