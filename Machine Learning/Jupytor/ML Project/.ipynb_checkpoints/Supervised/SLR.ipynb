SLR---(Simple Linear Regression)

```python
#importing Required Libraries

import numpy as np

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
# Defining the Variables and assigning values to them

temp = [20,25,30,35,40,45]#---- INDEPENDENT VARIABLE

icesale = [13,12,25,35,38,40] #---- DEPENDENT VARIABLE
```

```python
# Converting to arrays

X = np.array([temp]).T

Y = np.array(icesale)
```

```python
# Creating the model

model = LinearRegression()

# Fitting the model

model= model.fit(X,Y)
```

```python
# Getting the slope and intercept of the line

slope = model.coef_

intercept = model.intercept_
```

```python
# Prediction

predict = model.predict(X)

print(predict)
```

```python
# The plot shows the relationship between temperature and ice cream sales.

plt.scatter(temp, icesale, marker="*", edgecolors='r')
plt.plot(temp,predict, '-bo')
plt.show()
```
