KNN Method ------ (k-nearest neighbor algorithm)

```python

# Importing the necessary libraries

import pandas as pd
import numpy as np

```

```python
# Connecting and reading the database

database = pd.read_csv('diabetes.csv')
database.head

```

```python
# analyzing the data types and null values
# checking the data types and null values

database.info()
```

```python
# Removing null and zero values

no_zero = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

for i in no_zero:
    database[i] = database[i].replace(0, np.nan)
    mean= int(database[i].mean(skipna=True))
    database[i] = database[i].replace(np.nan, mean)

```

```python
database.head()
```

```python
#Importing required Models

from sklearn.model_selection import train_test_split #Splitting the data into train and test sets

from sklearn.preprocessing import StandardScaler   #Standardizing the data for future scaling

from sklearn.neighbors import KNeighborsClassifier # KNN Classifier for classification and initialization of Algorithm

from sklearn.metrics import accuracy_score #Accuracy score for the model
```

```python
# definig X and Y Axes

X = database.iloc[:, 0:9] # x axis values -- (0 to 8 columns)

Y = database.iloc[:, 8] # y axis values -- (8th column)

#Splitting the data into train and test sets and storing them in respective variables

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0) 
```

```python
#Future Scaling

sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
print(X_train)

```

```python
classifier = KNeighborsClassifier(n_neighbors=13, metric='euclidean')
```

```python
classifier.fit(X_train, Y_train)
```

```python
#Predicting the Test set results

Y_pred = classifier.predict(X_test)
print(Y_pred)
```

```python
#Calculating the accuracy ---- (GOOD IF THE ACCURACY IS GREATER THAN 70%)

accuracy_score(Y_test, Y_pred)
```
