K-Means Clustring (Used to find the cluster points of the dataset)

```python
# Importing Modules

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

from sklearn.cluster import KMeans
```

```python
# Importing the dataset

dataset = pd.read_csv('Mall_Customers.csv')

# Checking the dataset

print(dataset)
```

```python
# Checking the dataset  

X = dataset.iloc[:, [2, 3]].values

#print(X)
```

```python
#creating input list for elbow method

input_list = []
for i in range (1,20):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(X)
    input_list.append(kmeans.inertia_)
```

```python
print(input_list)
```

```python
# Plotting the elbow method

plt.figure(figsize=(10,5))

#giving style and colour of graph inputs

sns.set(context='paper',palette='Blues',style='whitegrid',font_scale=1,color_codes=True)

#Ploting the ghaph

sns.lineplot(x=range(1,20) , y=input_list,marker='o', color='blue')
```

```python
kmeans = KMeans(n_clusters = 5,init = 'k-means++')

y_means = kmeans.fit_predict(X)



print(y_means)
```

```python
# Plotting the clusters 

plt.figure(figsize=(12,7))


sns.scatterplot(x=X[y_means == 0, 0], y=X[y_means == 0, 1], s=50, c='Yellow', label='Cluster 1')
sns.scatterplot(x=X[y_means == 1, 0], y=X[y_means == 1, 1], s=50, c='blue', label='Cluster 2')
sns.scatterplot(x=X[y_means == 2, 0], y=X[y_means == 2, 1], s=50, c='green', label='Cluster 3')
sns.scatterplot(x=X[y_means == 3, 0], y=X[y_means == 3, 1], s=50, c='orange', label='Cluster 4')
sns.scatterplot(x=X[y_means == 4, 0], y=X[y_means == 4, 1], s=50, c='purple', label='Cluster 5')
```

```python
# Plotting the centroids


sns.scatterplot(x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')

#Plotting the grid

plt.grid(True)

#Giving title and labels

plt.title('Clusters of Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score (1-100)')
plt.legend()

# Printing the graph

plt.show()
```

```python

# Plotting both the clusters and their centroids

# Deciding the size of the figure

plt.figure(figsize=(12,7))

# Plotting the clusters

sns.scatterplot(x=X[y_means == 0, 0], y=X[y_means == 0, 1], s=50, c='Yellow', label='Cluster 1')
sns.scatterplot(x=X[y_means == 1, 0], y=X[y_means == 1, 1], s=50, c='blue', label='Cluster 2')
sns.scatterplot(x=X[y_means == 2, 0], y=X[y_means == 2, 1], s=50, c='green', label='Cluster 3')
sns.scatterplot(x=X[y_means == 3, 0], y=X[y_means == 3, 1], s=50, c='orange', label='Cluster 4')
sns.scatterplot(x=X[y_means == 4, 0], y=X[y_means == 4, 1], s=50, c='purple', label='Cluster 5')

# Plotting the centroids

sns.scatterplot(x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')
plt.grid(False)
plt.title('Clusters of Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
