DATA VISUALISATION

*Matplotlib*

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
prodid = [1, 2, 3, 4, 5]
sales = [1000, 200, 350, 600, 500]
```

```python
plt.plot(prodid,sales)
```

```python
plt.scatter(prodid, sales, color='red')
```

```python
df_prod = pd.DataFrame({"Prod_ID": prodid, "Sales": sales})
df_prod
```

```python
plt.scatter(df_prod["Prod_ID"], df_prod["Sales"])
```

```python
plt.scatter(df_prod["Prod_ID"], df_prod["Sales"])
plt.xlabel("Product ID")
plt.ylabel("Sales Amount")
plt.title("Sales by Product ID")
```

```python
clrs = ['r','b','g','c','y']
plt.scatter(df_prod["Prod_ID"], df_prod["Sales"],c=clrs)
plt.xlabel("Product ID")
plt.ylabel("Sales Amount")
plt.title("Sales by Product ID")
```

```python
clrs = ["r", "b", "g", "c", "y"]
size = [1000, 200, 300, 400, 500]
plt.scatter(df_prod["Prod_ID"], df_prod["Sales"], c=clrs, s=size)
plt.xlabel("Product ID")
plt.ylabel("Sales Amount")
plt.title("Sales by Product ID")
```

```python
clrs = ["r", "b", "g", "c", "y"]
size = [1000, 200, 300, 400, 500]
plt.scatter(df_prod["Prod_ID"], df_prod["Sales"], c=clrs, s=size,marker='*')
plt.xlabel("Product ID")
plt.ylabel("Sales Amount")
plt.title("Sales by Product ID")
```

```python
clrs = ["r", "b", "g", "c", "y"]
# size = [1000, 200, 300, 400, 500]
plt.bar(df_prod["Prod_ID"], df_prod["Sales"], color=clrs,width=0.5)
plt.xlabel("Product ID")
plt.ylabel("Sales Amount")
plt.title("Sales by Product ID")
```

```python
#alpha  is used to set the transparency of the bar
#alpha = 0.5 means 50% transparency
#alpha = 0.1 means 10% transparency
#alpha = 1 means 100% transparency (default value)
#alpha = 0 means 0% transparency (invisible)

clrs = ["r", "b", "g", "c", "y"]
plt.bar( df_prod["Prod_ID"], df_prod["Sales"], color=clrs, width=0.5, alpha=0.5)  # to see all parameters - shift+tab
plt.xlabel("Product IDs")
plt.ylabel("Sales Amount")
plt.title("ProdID vs SalesAmt")
```

```python
clrs = ["r", "b", "g", "c", "y"]
plt.barh( df_prod["Prod_ID"], df_prod["Sales"], color=clrs)  # to see all parameters - shift+tab
plt.ylabel("Product IDs")
plt.xlabel("Sales Amount")
plt.title("ProdID vs SalesAmt")
```

```python
plt.pie(df_prod["Sales"])
```

```python
highlighted=[1,0,0,0,1]
plt.pie(df_prod["Sales"],labels=df_prod["Prod_ID"],shadow=True,explode=highlighted,autopct='%1.1f%%')
```

```python
highlighted = [1, 0, 0, 0, 0]
plt.pie(df_prod["Sales"], labels=df_prod["Prod_ID"], shadow=True, explode=highlighted)
plt.legend(df_prod["Sales"])
plt.title("Pie Chart")
```

```python
highlighted = [1, 0, 0, 0, 0]
plt.pie(
    df_prod["Sales"],
    labels=df_prod["Prod_ID"],
    shadow=True,
    explode=highlighted,
    startangle=90,
)
plt.legend(df_prod["Sales"])
plt.title("Pie Chart")
```

```python
highlighted = [1, 0, 0, 0, 0]
plt.pie(
    df_prod["Sales"],
    labels=df_prod["Prod_ID"],
    shadow=True,
    explode=highlighted,
    startangle=90,
    labeldistance=1.5,
)
plt.legend(df_prod["Sales"])
plt.title("Pie Chart")
```
