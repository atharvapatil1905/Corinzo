DATA VISUALISATION

*Seaborn*

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prodid = [1, 2, 3, 4, 5]
sales = [1000, 200, 350, 600, 500]
```

```python
df_prod=pd.DataFrame({"Prod_ID":prodid,"Sales":sales})
df_prod
```

```python
sns.lineplot(x="Prod_ID", y="Sales", data=df_prod)
```

```python
sns.scatterplot(x="Prod_ID", y="Sales", data=df_prod)
```

```python
sns.countplot(x='Sales',data=df_prod)
```

SUBPLOT

```python
l1=[1,3,5,7]
l2=[10,30,90,50]
plt.plot(l1,l2)

l3=[10,3,5,70]
l4=[1,3,9,27]
plt.plot(l3,l4)
```

```python
l1=[1,3,5,7]
l2=[10,30,90,50]
plt.subplot(1,2,1) #no of rows,no of cols,pos
plt.plot(l1,l2)

l3=[10,3,5,70]
l4=[1,3,9,27]
plt.subplot(1,2,2) #no of rows,no of cols,pos
plt.plot(l3,l4)
```

```python
l1=[1,3,5,7]
l2=[10,30,90,50]
plt.subplot(2,1,1) #no of rows,no of cols,pos
plt.plot(l1,l2)

l3=[10,3,5,70]
l4=[1,3,9,27]
plt.subplot(2,1,2) #no of rows,no of cols,pos
plt.plot(l3,l4)
```
