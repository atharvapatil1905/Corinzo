House price prediction SVR

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import math
from sklearn.svm import SVR
```

```python
df = pd.read_csv(r"C:\Users\hp\repositories\Corizo\AI\House.csv")
df
```

```python
df["Bedrooms"].median()

```

```python
# The median is the middle value in a list of numbers. If the list has an even number of values, the median is the average of the two middle numbers.
# The median is a measure of central tendency that is less affected by outliers than the mean.
med_bed = math.floor(df["Bedrooms"].median())
print(med_bed)
```

```python
df.interpolate()
```

```python
df["Bedrooms"].fillna(med_bed, inplace=True)
df
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    df[["Area", "Bedrooms", "Age"]], df["Price"], train_size=0.8
)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
# Feature Scaling
SVR_model = SVR()
SVR_model.fit(xtrain, ytrain)
```

```python
ans = SVR_model.predict(xtest)
ans
```

```python
SVR_model.score(xtest, ytest)
```
