```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df3=pd.read_csv(r"C:\Users\hp\repositories\Corizo\AI\Salaries.csv")
df3
```

```python
# Predict Salaries for Level 11 (director)
# Salaries - Output
# Position,Level - Input, Level - updated input
```

```python
df3.isnull().sum()
```

```python
# x-input
x = df3[["Level"]]
x
# y-output
```

```python
y = df3[["Salaries"]]
y
```

```python
from sklearn.linear_model import LinearRegression

model3 = LinearRegression()
```

```python
model3.fit(x,y)
```

```python
model3.predict([[11]])  # using linear regression
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_model = PolynomialFeatures(degree=8)  # degree=2 ...n
```

```python
x
```

```python
x_poly = poly_model.fit_transform(x)
x_poly
```

```python
from sklearn.linear_model import LinearRegression

model4=LinearRegression()
```

```python
model4.fit(x_poly, y)
```

```python
lev = 11
trans_lev = poly_model.fit_transform([[11]])
trans_lev
```

```python
model4.predict(trans_lev)  # degree 2 polynomial regression
```

```python
plt.scatter(x, y, color="b")
plt.plot(x, model3.predict(x), color="r")
plt.xlabel("LEVEL")
plt.ylabel("SALARY")
plt.title("Using Linear Regression")
```

```python
plt.scatter(x, y, color="b")
plt.plot(x, model4.predict(poly_model.fit_transform(x)), color="r")
plt.xlabel("LEVEL")
plt.ylabel("SALARY")
plt.title("Using Polynomial Regression")
```
