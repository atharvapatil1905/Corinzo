Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"C:\Users\hp\repositories\Corizo\AI\Insurance_LR.csv")
df

# Predict Premium Price for age - 37,83,67
```

```python
plt.scatter(df["Age"],df["Premium"])
plt.xlabel("Age")
plt.ylabel("Premium")
```

```python
sns.lmplot(x="Age", y="Premium",data=df)
```

```python
print(type(df["Age"]))
print(type(df["Premium"]))
```

```python
model = LinearRegression()  # y=mx+c
model.fit(df[["Age"]], df[["Premium"]])
```

```python
print(model.predict([[37]]))
print(model.predict([[83]]))
print(model.predict([[67]]))
```

```python
# y=mx+c , y=premium, x=age, m=slope/gradient, c=intercept
x = 37
m = model.coef_
print(m)
c = model.intercept_
print(c)
y = m * x + c
print(f"Premium Price: {y} for Age: {x}")
```
