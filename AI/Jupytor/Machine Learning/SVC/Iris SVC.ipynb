```python
import pandas as pdd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv(r'C:\Users\hp\repositories\Corizo\AI\Iris.csv')
df
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=['Species']), df['Species'], test_size=0.2, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.svm import SVC  # Support Vector Classifier

# Create a Support Vector Classifier model
# and fit it to the training data
svc_model = SVC()
svc_model.fit(x_train, y_train)
```

```python
ans=svc_model.predict(x_test)
```

```python
svc_model.score(x_test, y_test)
```

```python
svc_model.predict([[103,4.9, 3.0, 1.4, 0.2]]) # Iris Setosa
```

```python
from sklearn.metrics import confusion_matrix


# Confusion matrix
cm = confusion_matrix(y_test, ans)
cm
```

```python
sns.heatmap(
    cm,
    annot=True,
    xticklabels=["Iris-setosa", "Iris-versicolor", "Iris-virginica"],
    yticklabels=["Iris-setosa", "Iris-versicolor", "Iris-virginica"],
)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Confusion Matrix")
plt.show()
```
